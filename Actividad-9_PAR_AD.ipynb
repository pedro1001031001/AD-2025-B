<h3> Modelo con KNN</h3>
<p>
    El modelo K-Nearest Neighbors es un modelo predictivo clasificado dentro del aprendizaje supervisado no paramétrico, en el cual usa la proximidad para hacer clasificaciones o predicciones sobre la agrupación de un punto de datos individual (Núñez Sánchez, 2023, p. 7); es decir, clasifica un objeto en función de de los objetos etiquetados más cercanos con el conjunto de datos (Jiang, 2023, p. 68).<br>
    &emsp; Se busca estimar el valor pérdido, como: la media de los valores de los k vecinos u obsevaciones más cercanas (Núñez Sánchez, 2023, p. 7). Para llevar a cabo el cálculo de la cercanía de los K vecinos, se usa distintas medias: Distancia Euclidiana o de Manhattan (Núñez Sánchez, 2023, p. 7). <br>
    &emsp; El valor de K representa el número de vecinos más cercanos (Jiang, 2023, p. 68; Núñez Sánchez, 2023, p. 8).<br>
    &emsp; Para implementar este modelo, se propone la siguiente metodología:
    <ol>
        <li> Procesamiento de los datos.</li>
        <li> Exploración de los datos.</li>
        <li> Representación de los datos.</li>
        <li> Cálculo de la distancia.</li>
        <li> Selección de los vecinos más cercanos.</li>
        <li> Clasificación o predicción.</li>
        <li> Décision (Richaud, 2024, p. 3).</li> <br>
    </ol>
    A continuación, se realizará una demostración del modelo.
</p>


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score

import random

# 1. Creación de datos
m = 100
car = 5
clas = 4

x = np.random.rand(m, car)
y = np.random.randint(1, clas, size= m)

# 2. División de los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)

# 3. Creación y entrenamiento del modelo
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)

# 4. Predicciones
predicciones = knn.predict(X_test)
print(f"Predicciones: {predicciones}")

accuracy = accuracy_score(y_test, predicciones)
print(f"Precisión del modelo (multiclase): {accuracy:.2f}")

# 5. Reducción de dimensiones para visualizar
pca = PCA(n_components=2)
X_test_pca = pca.fit_transform(X_test)

# 6. Visualización
    # Graficar datos de entrenamiento
plt.figure(figsize=(10, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='coolwarm', label='Entrenamiento', edgecolors='k', alpha=0.7)

    # Graficar datos de prueba con predicciones
plt.scatter(X_test[:, 0], X_test_pca[:, 1], c=predicciones, cmap='coolwarm', marker='x', s=100, label='Predicciones (Prueba)')

plt.title("Predicciones KNN con múltiples clases (reducidas a 2D con PCA)")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.legend()
plt.grid(True)
plt.show()
```

<p>
    En la anterior demostración, se llevo a cabo las siguientes actividades:
    <ol>
        <li> Primero, se creó los datos aleatorios de tipo numérico.</li>
        <li> Seguido, se separó los datos en conjuntos de entrenamiento y de prueba, para alimentar al modelo.</li>
        <li> Siguiente, se entrenó el modelo.</li>
        <li> Luego, se realizaron las predicciones para reconocer cómo el modelo puede clasificar nuevas instancias. Y se evaluó las predicicones</li>
        <li> Ademas, se redujó la dimensión a 2D.</li>
        <li> Y, por último, se graficaron los predicciones y los datos de entrenamiento.</li>
    </ol>
</p>

<h3> Modelo con K-means</h3>
<p {line-height: 1.5;}>
    El <i>Machine Learning</i> es una rama de la <i>IA</i>, la cual se enfoca en el estudio de los algoritmos para mejorar desempeño y eficiencia de los programas (Farhan AiShammari, 2024, p. 12). Y, dentro de las aplicaciones del <i>Machine Learning</i> está el <i>Clustering</i>, el cual es un algoritmo de aprendizaje no supervisado, donde los datos son divididos en grupos, de acuerdo con sus caracteríticas (Farhan AiShammari, 2024, p. 12); por otra parte, para ser aplicado, se considera el usa del método de K-means (Farhan AiShammari, 2024, p. 12). <br>    
    &emsp; K-means es un método matemático usado para dividir los datos en grupos usando las medias (o los <i>centers</i>) (Farhan AiShammari, 2024, p. 13). Este método es iterativo, el cual se mantiene en constante ejecución hasta obtener los clusters finales (Farhan AiShammari, 2024, p. 12). Durante su ejecución, este método es usado para medir y organizar las distancias entre los puntos de datos y los <i>centers</i>; donde los puntos de datos son asignados a los clusters más cercanos a los <i>centers</i> (Farhan AiShammari, 2024, p. 12). <br>
    &emsp; Para llevar a cabo el <i>Clustering</i>, se propone que se lleve a cabo la siguiente metodología:
    <ol>
        <li> Preparación de los datos. </li>
        <li> Inicialización de los <i>centers</i>. </li>
        <li> "Computing labels": </li>
            <ol>
                <li> "Computing distances". </li>
                <li> "Finding minimum distance". </li>
                <li> "Assigning labels". </li>
            </ol>
        <li> "Computing clusters". </li>
        <li> "Computing error function". </li>
        <li> Actualización de los <i>centers</i>. </li>
        <li> Ilustración de los <i>clusters</i> (Farhan AiShammari, 2024, p. 14). </li> <br>
    </ol>
    &emsp; A continuación, se implementará el modelo de K-means.
</p>

```python
# Librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import math

# 1. Preparación de los datos.
x = np.random.rand(500).astype(float)
y = np.random.rand(500).astype(float)

k = 5 # Número de clusters
    
data = [ (float(xi), float(yi)) for xi, yi in zip(x, y) ] # dataframe

# 2. Inicialización de los centros.
def inicializar_centers(data, k): 
    centers = []
    
    for i in range(k):
        index = random.randint(0, len(data) - 1)
        centers.append(data[index])
    return centers

# 3. "Computing labels."

    # 3.1 Computing distances
def euclidean(point1, point2):
    sum = 0

    for i in range(len(point1)):
        sum += (point1[i] - point2[i])**2
    return math.sqrt(sum)

    # 3.2 Calcular las distancias de los puntos
def compute_distances(point, centers):
    distances = []
    
    for i in range(len(centers)):
        distance = euclidean(point, centers[i])
        distances.append(distance)
    return distances

    # 3.3 Etiquetas
def compute_labels(data, centers):
    labels = []
    
    for i in range(len(data)):
        distances = compute_distances(data[i], centers)
        min_distance = min(distances)
        label = distances.index(min_distance)
        labels.append(label)
    return labels

# 4. "Computing clusters."
def compute_clusters(data, labels, k):
    clusters = []
    
    for i in range(k):
        clusters.append([])
        
    for i in range(len(data)):
        label = labels[i]
        clusters[label].append(data[i])
    return clusters
            
# 5. "Computing error function"
def compute_error(clusers, centers):
    sum = 0
    
    for i in range(len(clusters)):
        for j in range(len(clusters[i])):
            sum += euclidean(clusters[i][j], centers[i])**2
    return sum

# 6. Actualización de los "centers"
def update_centers(clusters):
    centers = []
    
    for i in range(len(clusters)):
        sumx = 0
        sumy = 0
        
        for j in range(len(clusters[i])):
            sumx += clusters[i][j][0]
            sumy += clusters[i][j][1]
        
        centers.append([sumx/len(clusters[i]), sumy/len(clusters[i])])
    return centers

# Ejecución _______________________________________________________________________
centers = inicializar_centers(data, k)

nt = 500
i = 1

for i in range(nt):
    labels = compute_labels(data, centers)
    clusters = compute_clusters(data, labels, k)
    error = compute_error(clusters, centers)
    old_centers = centers
    centers = update_centers(clusters)
    
    if centers == old_centers:
        break
        
# 7. Ilustración de los "clusters"
plt.figure(figsize=(10, 6))
plt.scatter(x, y)
for i in range(len(clusters)):
    cluster_x = np.transpose(clusters[i])[0]
    cluster_y = np.transpose(clusters[i])[1]
    plt.scatter(cluster_x, cluster_y, label=f'Cluster {i+1}')

centers = np.array(centers)
plt.scatter(centers[:, 0], centers[:, 1], c='black', marker='*', s=100, linewidths=2, label='Centers')

plt.title("Clusters generados con K-means")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show
```

<p>
    En la aplicación del modelo anterior, se ejecutó las siguientes actividades:
    <ol>
        <li> Primero, se creó los datos a emplear.</li>
        <li> Seguido, se creó la función que inicializaría los centros.</li>
        <li> Después, se preparó las funciones para medir distancias y asignar etiquetas.</li>
        <li> Luego, se creó la función para crear los clusters, de acuerdo con las etiquetas.</li>
        <li> Además, se implentó la función de errores.</li>
        <li> A continuación, se creó la función que actualizará a los centros.</li>
        <li> Por último, se ilustra los centros y datos agrupados.</li>
    </ol>
